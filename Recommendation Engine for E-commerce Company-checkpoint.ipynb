```python
import pandas as pd
import numpy as np
```

```python
rating_df = pd.read_csv( "CF_Input5.csv")
```

```python
rating_df.head()
```

```python
#Finding User Similarities

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#User_item Metrics

user_items_df = rating_df.pivot( index='user', columns='item', values = "rating" ).reset_index(drop=True)

#Fill '0' for ratings not given by users

user_items_df.fillna( 0, inplace = True )

user_items_df.head()
```

```python
#item_user Metrics

item_user_df = rating_df.pivot( index='item', columns='user', values = "rating" ).reset_index(drop=True)

#Fill '0' for ratings not given by users

item_user_df.fillna( 0, inplace = True )

item_user_df.head()
```

```python
usernames = item_user_df.columns
```

```python
usernames
```

```python
itemnames = user_items_df.columns
```

```python
itemnames
```

```python
user_sim = 1 - pairwise_distances( user_items_df.as_matrix(), metric= 'cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame( user_sim, index= usernames, columns=usernames)
```

```python
user_sim_df.head()
```

```python
item_sim = 1 - pairwise_distances( item_user_df.as_matrix(), metric= 'cosine')
```

```python
item_sim_df = pd.DataFrame( item_sim, index= itemnames, columns=itemnames )
```

```python
item_sim_df.head()
```

```python
user_sim_df['USER_00001']
```

```python
users_df = pd.DataFrame()
users_df['similarity'] = user_sim_df['USER_00001']
```

```python
users_df.sort_values( ["similarity"], ascending = False )[1:6]
```

```python
def get_similar_users( userid, topN = 5 ):
    users_df = pd.DataFrame()
    users_df['similarity'] = user_sim_df[userid]
    top_n = users_df.sort_values( ["similarity"], ascending = False )[1:topN+1]
    print( "Top 5" + " Similar users to: " + userid + " are", )
    return top_n
```

```python
get_similar_users('USER_00003')
```

```python
def get_similar_items( itemid, topN = 5 ):
    items_df = pd.DataFrame()
    items_df['similarity'] = item_sim_df[itemid]
    top_n = items_df.sort_values( ["similarity"], ascending = False )[1:topN+1]
    print( "Top 5 Similar items to: " + itemid + " are", )
    return top_n

get_similar_items( 'IT_06' )
```

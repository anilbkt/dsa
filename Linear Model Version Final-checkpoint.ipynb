### Step 2.3 : Data Preparation - Outlier Treatment, Missing Value Treatment, Dummy Variable Creation etc.

```python
# Handling Outliers - Using quantile method. Setting upper limit at 99 quantile and lowe at 1 quantile. 
# (To be applied only on the numerical variables)

# This is not a refined process
# Please fine tune it!!
# Look at the symmetry

def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x


data_num=data_num.apply(outlier_capping)
```

```python
# Handling Missing values - Using Median imputation method.
# (To be applied only on the numerical variables)


def Missing_imputation(x):
    x = x.fillna(x.median())
    return x


data_num=data_num.apply(lambda x: Missing_imputation(x))
```

```python
data_num.shape[0]-data_num.count()
```

```python
# Handling Missing values - Using Mode imputation method.
# (To be applied only on the categorical variables)


def Cat_Missing_imputation(x):
    x = x.fillna(x.mode())
    return x

data_cat=data_cat.apply(lambda x: Cat_Missing_imputation(x))
```

```python
# For all cat variables we are creating dummy variables
# 1. Is it a best practise ? No
# 2. Why??
#          - some variables are ordinal and don't require dummies
#          - NOT ALL Cat vars ARE IMPORTANT FOR Y
# What is the best practice
#  - Perform this stem after performing ANOVA

cols = data_cat.columns.to_list()
data_cat = pd.get_dummies(columns=cols,data = data_cat,drop_first=True)
```
